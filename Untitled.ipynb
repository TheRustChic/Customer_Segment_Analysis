```python
df = pd.read_csv("Mall_Customers_extracted.csv")
df.head()
```

```python
df['Genre'] = df['Genre'].map({'Male': 0, 'Female': 1})
```

```python
df.head()
```

```python
df_clean = df.drop(columns=['Genre', 'CustomerID'])  # assuming 'CustomerID' is non-numeric
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clean)
```

```python
from sklearn.cluster import KMeans

# Set number of clusters
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis')
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Annual_Income_(k$)'], df['Spending_Score'], c=df['Cluster'], cmap='viridis')
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.colorbar(label='Cluster')
plt.savefig("customer_segments_i.png")  # <- this line saves it
plt.show()

```

```python
df['Cluster'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Customer Distribution by Cluster')
plt.ylabel('')
plt.savefig("pie_chart.png")
plt.show()
```

```python
import seaborn as sns
sns.barplot(x='Cluster', y='Annual_Income_(k$)', data=df)
plt.title('Average Income by Cluster')
plt.savefig("income_by_cluster.png")
plt.show()

sns.barplot(x='Cluster', y='Spending_Score', data=df)
plt.title('Average Spending Score by Cluster')
plt.savefig("spending_by_cluster.png")
plt.show()
```

```python
df.groupby('Cluster').mean()
# Group by Segment and calculate mean only for numeric columns
cluster_summary = df.groupby('Segment').mean(numeric_only=True)
cluster_summary.to_csv("cluster_summary.csv")


```

```python
cluster_map = {
    0: 'Older Moderate Spenders',
    1: 'VIP Customers',
    2: 'Young High Spenders',
    3: 'High Income Low Spenders'
}

df['Segment'] = df['Cluster'].map(cluster_map)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Annual_Income_(k$)',
    y='Spending_Score',
    hue='Segment',
    palette='Set2',
    s=100
)
plt.title('Customer Segments by Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Customer Segment')
plt.grid(True)
plt.savefig("customer_segments.png")  # <- this line saves it
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Segment', palette='Set2')
plt.xticks(rotation=20)
plt.title('Customer Count per Segment')
plt.ylabel('Count')
plt.xlabel('Segment')
plt.grid(axis='y')
plt.savefig("customer_count.png")  # <- this line saves it
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Segment', y='Age', palette='Set2')
plt.title('Age Distribution by Customer Segment')
plt.ylabel('Age')
plt.xlabel('Segment')
plt.xticks(rotation=20)
plt.grid(axis='y')
plt.savefig("customer_age.png")  # <- this line saves it
plt.show()
```
